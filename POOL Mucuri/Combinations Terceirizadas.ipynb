```python
import pandas as pd
import os
```

```python
arquivos_terceirizadas = r"C:\Users\MarkosPereira.exc\Downloads\BIdopool\ZGeral\Terceirizadas".replace("\\","/")
```

```python
for raiz, diretorios, arquivos in os.walk(arquivos_terceirizadas):
    for arquivo in arquivos:
        print(arquivo)
```

```python
#Transformando os arquivos em fata frames individuais e adicionando-os em uma lista
df_arq_cap = list()
for raiz, diretorios, arquivos in os.walk(arquivos_terceirizadas):
    for arquivo in arquivos:
        caminho_completo = os.path.join(raiz, arquivo)
        df = pd.read_excel(caminho_completo, engine='openpyxl', parse_dates=True)
        df_arq_cap.append(df)
```

```python
for dataframe in df_arq_cap:
    display(dataframe.head(6))
```

```python
#As colunas não estão padronizadas, nesse caso faz se alterações na base de jacarei
#renomeia-se a coluna "C/Custo" por "C. CUSTO"
#renomeia-se a coluna "VL Total" por "VALOR TOTAL"
```

```python
for dataframe in df_arq_cap:
    print(dataframe.columns)
    print(dataframe.shape)
```

```python
df_arq_cap[1] = df_arq_cap[1].rename(columns={'C.CUSTO': 'C. CUSTO'})
df_arq_cap[1] = df_arq_cap[1].rename(columns={'CUSTO': 'VALOR TOTAL'})
df_arq_cap[1] = df_arq_cap[1].rename(columns={'CAP': 'ST'})
df_arq_cap[1] = df_arq_cap[1].rename(columns={'SOLICTANTE ': 'SOLICITANTE'})
df_arq_cap[1] = df_arq_cap[1].rename(columns={'tipo ': 'TIPO'})

df_arq_cap[4] = df_arq_cap[4].rename(columns={'C/CUSTO': 'C. CUSTO'})
df_arq_cap[4] = df_arq_cap[4].rename(columns={'VL TOTAL': 'VALOR TOTAL'})
```

```python
for dataframe in df_arq_cap:
    print(dataframe.columns)
```

```python
data_frame_terceirizadas = pd.concat(df_arq_cap)
```
