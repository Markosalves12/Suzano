```python
import pandas as pd
import os
pd.options.display.max_columns = 50
```

```python
arquivos_uber = r"C:\Users\MarkosPereira.exc\Suzano S A\BI Mobilidade unificado\arquivos uber".replace("\\","/")
```

```python
for raiz, diretorios, arquivos in os.walk(arquivos_uber):
    for arquivo in arquivos:
        print(arquivo)
```

```python
#Transformando os arquivos em fata frames individuais e adicionando-os em uma lista
df_arq_cap = list()
for raiz, diretorios, arquivos in os.walk(arquivos_uber):
    for arquivo in arquivos:
        caminho_completo = os.path.join(raiz, arquivo)
        df = pd.read_excel(caminho_completo, engine='openpyxl', parse_dates=True, header=5)
        df_arq_cap.append(df)
```

```python
for dataframe in df_arq_cap:
    display(dataframe.head(6))
```

```python
for dataframe in df_arq_cap:
    print(dataframe.columns)
    print(dataframe.shape)
```

```python
columns_ultils = {
                    'data':'Data da solicitação (local)',
                    'hora':'Hora da solicitação (local)',
                    'origem':'Endereço de partida',
                    'destino':'Endereço de destino',
                    'valor':'Valor da transação em BRL (com tributos)',
                    'solicitante1':'Nome',
                    'solicitante2':'Sobrenome',
                    'passageiro1':'Nome',
                    'passageiro2':'Sobrenome',
                    'centro_de_custo':'CC',
                 }
```

```python
ultils_data_frames = list()
for dataframe in df_arq_cap:
    dict_to_convert = dict()
    for key, value in columns_ultils.items():
        try:
            dict_to_convert[key] = dataframe[value]
        except:
            continue
    ultil_data_frame = pd.DataFrame(dict_to_convert)
    for coluna in ultil_data_frame.columns:
        ultil_data_frame[coluna] = ultil_data_frame[coluna].astype(str)
    ultils_data_frames.append(ultil_data_frame)
```

```python
for dataframe in ultils_data_frames:
    display(dataframe)
```

```python
for dataframe in ultils_data_frames:
    dataframe['solicitante'] = dataframe['solicitante1']+dataframe['solicitante2']
    dataframe['passageiro'] = dataframe['passageiro1']+dataframe['passageiro2']
    dataframe = dataframe.drop(columns=['solicitante1', 'solicitante2', 'passageiro1', 'passageiro2'])
    display(dataframe)
```

```python
finals_df = list()
for dataframe in ultils_data_frames:
    if dataframe.shape[1] == 11:
        dataframe = dataframe[['data', 'hora', 'origem', 'destino', 'valor', 'solicitante1',
          'solicitante2', 'passageiro1', 'passageiro2', 'solicitante',
          'passageiro']]
    else:
        dataframe = dataframe[['data', 'hora', 'origem', 'destino', 'valor', 'solicitante1',
          'solicitante2', 'passageiro1', 'passageiro2', 'solicitante',
          'passageiro', 'centro_de_custo']]
        
    finals_df.append(dataframe)
```

```python
for dataframe in finals_df:
    display(dataframe)
```

```python
caminho_uber = r"C:\Users\MarkosPereira.exc\Suzano S A\BI Mobilidade unificado\Total_chamados_uber.xlsx".replace("\\","/")
uber_combinado = pd.concat(finals_df)
uber_combinado = uber_combinado.assign(tabela='Uber')
```

```python
uber_combinado = uber_combinado.drop(columns=['solicitante1', 'solicitante2', 'passageiro1', 'passageiro2'], axis=1)
```

```python
uber_combinado.to_excel(caminho_uber)
```
