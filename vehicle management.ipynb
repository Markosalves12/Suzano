```python
import pandas as pd

def convert_dict_to_table(data_dict):
    # Extrair os valores desejados de 'IOData' 'Inputs'
    inputs = data_dict.get('IOData', {}).get('Inputs', [])
    input_dict = {entry['Name']: entry['Value'] for entry in inputs}
    
    # Extrair os valores desejados de 'IOData' 'Outputs'
    outputs = data_dict.get('IOData', {}).get('Outputs', [])
    output_dict = {entry['Name']: entry['Value'] for entry in outputs}
    
    # Criar um DataFrame pandas
    df = pd.DataFrame({
        'ExternalId': [data_dict.get('ExternalId')],
        'Direction': [data_dict.get('Direction')],
        'AccountID': [data_dict.get('AccountID')],
        'GroupId': [data_dict.get('GroupId')],
        'AccountName': [data_dict.get('AccountName')],
        'UnitID': [data_dict.get('UnitID')],
        'LicensePlate': [data_dict.get('LicensePlate')],
        'FleetAlias': [data_dict.get('FleetAlias')],
        'FleetNumber': [data_dict.get('FleetNumber')],
        'FuelType': [data_dict.get('FuelType')],
        'Manufacturer': [data_dict.get('Manufacturer')],
        'Model': [data_dict.get('Model')],
        'Year': [data_dict.get('Year')],
        'VehicleID': [data_dict.get('VehicleID')],
        'LastMessageTime': [data_dict.get('LastMessageTime')],
        'LastLocationTime': [data_dict.get('LastLocationTime')],
        'Lat': [data_dict.get('Lat')],
        'Long': [data_dict.get('Long')],
        'Address': [data_dict.get('Address')],
        'Speed': [data_dict.get('Speed')],
        'Battery': [data_dict.get('Battery')],
        'Ignition': [data_dict.get('Ignition')],
        'DriverName': [data_dict.get('DriverName')],
        'Odometer': [data_dict.get('Odometer')],
        'Ignio': [input_dict.get('Ignio', None)],
        'Limpador do Para-brisa': [input_dict.get('Limpador do Para-brisa', None)],
        'Sirene': [output_dict.get('Sirene', None)]
    })
    
    return df
```

```python
df_arq_cap = list()

for key in y['Result'].keys():
    element = y['Result'][key]
    
    for value in element:
        df = convert_dict_to_table(value)
        df = df.assign(origin=key)
        df_arq_cap.append(df)
```

```python
def dividir_e_salvar_dataframe(df, diretorio, string):
    # Verifique o tamanho do DataFrame
    num_linhas = len(df)

    if num_linhas > 1000000:
        # Divida o DataFrame em partes menores de 1 milho de linhas
        num_partes = (num_linhas // 1000000) + 1

        for i, chunk in enumerate(np.array_split(df, num_partes)):
            # Crie o nome do arquivo com base na string e no ndice
            nome_arquivo = f"{string}_{i + 1}.xlsx"
            caminho_arquivo = os.path.join(diretorio, nome_arquivo)

            # Salve o chunk como um arquivo CSV com os cabealhos
            chunk.to_excel(caminho_arquivo, index=False)

    else:
        # Se o DataFrame for pequeno, salve-o como um nico arquivo
        nome_arquivo = f"{string}.xlsx"
        caminho_arquivo = os.path.join(diretorio, nome_arquivo)
        df.to_excel(caminho_arquivo, index=False)
```

```python
def lista_de_datas(data_inicio, data_final):
    # Converte as datas de entrada para objetos datetime
    data_inicio = datetime.strptime(data_inicio, '%Y/%m/%d')
    data_final = datetime.strptime(data_final, '%Y/%m/%d')

    # Inicializa a lista de datas
    datas = []

    # Adiciona as datas  lista enquanto a data de incio for menor ou igual  data final
    while data_inicio <= data_final:
        datas.append(data_inicio.strftime('%Y/%m/%d'))
        data_inicio += timedelta(days=1)

    return datas
```

```python
def convert_history_rout(values):
    DriverId = list()
    DateTime = list()
    TxReason = list()
    GroupId = list()
    UnitID = list()
    LicensePlate = list()
    VehicleID = list()
    Lat = list()
    Long = list()
    Address = list()
    Speed = list()
    Ignition = list()
    DriverName = list()


    for value in values:
        DriverId.append(value['DriverId'])
        DateTime.append(value['DateTime'])
        TxReason.append(value['TxReason'])
        GroupId.append(value['GroupId'])
        UnitID.append(value['UnitID'])
        LicensePlate.append(value['LicensePlate'])
        VehicleID.append(value['VehicleID'])
        Lat.append(value['Lat'])
        Long.append(value['Long'])
        Address.append(value['Address'])
        Speed.append(value['Speed'])
        Ignition.append(value['Ignition'])
        DriverName.append(value['DriverName'])


    GetVehicleRouteHistoryInfo = pd.DataFrame({'DriverId': DriverId, 'DateTime': DateTime, 'TxReason': TxReason,
                                              'GroupId': GroupId, 'UnitID': UnitID, 'LicensePlate': LicensePlate,
                                              'VehicleID': VehicleID, 'Lat': Lat, 'Long': Long, 'Address': Address,
                                              'Speed': Speed, 'Ignition': Ignition,
                                               'DriverName': DriverName}).drop_duplicates()

    GetVehicleRouteHistoryInfo = GetVehicleRouteHistoryInfo[(GetVehicleRouteHistoryInfo['Lat']!=0)
                                                            &(GetVehicleRouteHistoryInfo['Long']!=0)]
    
    return GetVehicleRouteHistoryInfo
        
```
